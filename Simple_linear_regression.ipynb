# Train and Test Dataset

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(cdf, test_size=0.2, random_state=42)
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='palevioletred')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

# Modeling

```python
from sklearn import linear_model

# we build our model
regr = linear_model.LinearRegression()

# We specify our training data
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])

# Here we fit our model :)
regr.fit(train_x , train_y)


print("We built our model here's the coefficient and Intercept")
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

# Plot output

```python



# Plot the data points
plt.scatter(train_x, train_y, color='hotpink', label='Actual Data')

# Plot the regression line
plt.plot(train_x, regr.predict(train_x.reshape(-1, 1)), color='blue', linewidth=3, label='Regression Line')

# Set labels and title
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.title('Linear Regression Plot')

# Display legend
plt.legend()

# Show the plot
plt.show()
```

# Model Evaluation

```python
from sklearn.metrics import r2_score

# specifying our test data
test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])

# predicting y_test
test_y_predict = regr.predict(test_x)

print('Mean of actual test_y: {}'.format(np.mean(test_y)))
print('Mean of predicted test_y: {}'.format(np.mean(test_y_predict)))


print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_predict - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_predict - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y , test_y_predict) )
```
